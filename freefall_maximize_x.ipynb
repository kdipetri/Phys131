```python
import math
```

```python
def y(v0,theta,t):
    y = v0 * math.sin(theta) * t - 0.5 * 10 * t**2
    return y

def x(v0,theta,t):
    x = v0 * math.cos(theta) * t
    return x
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-colorblind')

# fix velocity
v0 = 20 # m/s

# we'll use this to get x and y versus time
ts = np.linspace(0,10,101) # s
#print(ts)

# initial angles to test
thetas = np.linspace(0,90,7) # deg
#print(thetas)

# plot x versus y for different initial theta
fig, ax = plt.subplots()

for th in thetas: 
    
    theta = th * math.pi/180 # convert to rad
    xs = [x(v0,theta,t) for t in ts if y(v0,theta,t)>0] # x verus t, but only if y>0
    ys = [y(v0,theta,t) for t in ts if y(v0,theta,t)>0] # y verus t, but only if y>0

    ax.plot(xs,ys,label=r'$\theta={:.0f}^\circ$'.format(th))

ax.legend()
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
```

```python
# trying interactive! 

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
%matplotlib inline

v0=20 # m/s

def plot_func(theta):
    
    theta = theta * math.pi/180. # convert to rad
    
    tmax = v0/5. * math.sin(theta)
    
    ts = np.linspace(0,tmax+1,10000) # s
    xs = [x(v0,theta,t) for t in ts if y(v0,theta,t)>=0] # x verus t, but only if y>0
    ys = [y(v0,theta,t) for t in ts if y(v0,theta,t)>=0] # y verus t, but only if y>0
    
    
    plt.plot(xs, ys)
    plt.xlabel('x [m]')
    plt.ylabel('y [m]')
    plt.show()
    

interact(plot_func, theta = widgets.FloatSlider(value=30,
                             min=0,
                             max=90,
                             step=1) )

```
